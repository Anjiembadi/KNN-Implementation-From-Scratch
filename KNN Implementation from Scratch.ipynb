## Step-2: Identify Input (X) and Output (y)

- Target : Price
- input variable: all columns expect price

```python
X=data.drop('price',axis=1)
y=data['price']
X.shape,y.shape
```

## Step-3: Train-Test Split (75 : 25)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## Step-4: Data Preprocessing on X_train

```python
from sklearn.preprocessing import OrdinalEncoder
```

#### Categorical Data Encoding

```python
#ordinal order
#cut
cut_order=['Fair','Good','Very Good','Premium','Ideal']
#color
color_order=['J','I','H','G','F','E','D']
#clarity
clarity_order=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

#### Ordinal Encoding

```python
oe=OrdinalEncoder(categories=[cut_order,color_order,clarity_order])
oe
```

#### Numerical Data Rescaling (RobustScaler)

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler=RobustScaler()
scaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer(
    [('ORDINAL',oe,['cut','color','clarity']),
    ('SCALE',scaler,['carat','depth','table','x','y','z'])],
    verbose_feature_names_out=False,
    remainder='passthrough',).set_output(transform='pandas')
ct
```

```python
X_train_transformed=ct.fit_transform(X_train)
X_train_transformed.head()
```

## Step - 5: Data Preprocessing on X_test 

```python
X_test_transformed=ct.transform(X_test)
X_test_transformed.head()
```

## Step - 6: Build the model and predict on X_test

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
query=X_test_transformed.sample(n=1,random_state=42)
query
```

```python
dist=pd.DataFrame(euclidean_distances(X_train_transformed,query)).sort_values(by=0,ascending=True)
dist.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knnr = KNeighborsRegressor(n_neighbors=5)
knnr
```

```python
knnr.fit(X_train_transformed, y_train)
```

```python
knnr.predict(query)
```

```python
y_predict=knnr.predict(X_test_transformed)
y_predict
```

## Step - 7: Evaluate your model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_predict)
mae
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
rmse
```

```python
r2 = r2_score(y_test, y_predict)
r2
```
